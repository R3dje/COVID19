```python
import pandas as pd
import numpy as np
from scipy import special

%matplotlib inline
```

```python
df = pd.DataFrame({'x': range(200), 'y': range(100, -100, -1)})
```

```python
df['PB'] = ((df['x'] ** 2) / -2 + 100 * df['x']) / 50
```

```python
df['EX'] = (10 ** (df['PB'] / 20)) / 1000
```

```python
df['EF'] = special.erf(df['x'] / 50) * 100
```

```python
df['LG'] = np.log10(df['x'] * 50) * 25
```

```python
df.plot()
```
